```python
import matplotlib.pyplot as plt
import seaborn as sns
# Use Pandas to Load the Dataabs
import pandas as pd
#load Csv File
df = pd.read_csv("students_performance.csv")
df
```

1. Load & Inspect the Dataset

```python
# Check shape and column names
print("Shape :", df.shape)
print("Column Name :", df.columns.tolist())
# Check for missing values
df.isnull().sum()
# Summary statistics
df.describe(include='all')
```

2. Univariate Analysis

```python
# Plot histograms
df[['math score', 'reading score' , 'writing score']].hist()
plt.title('Distribution of Scores')
plt.show()
```

```python
#Count Plots for Categorical data

Categorical_col = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']
for col in categorical_col:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col)
    plt.title(f'Distribution of {col}')
    plt.show()
    
```

3. Bivariate Analysis

```python
# Groupby gender
df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
score_cols = ['math score', 'reading score', 'writing score']
for col in score_cols:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x='gender', y=col, data=df)
    plt.title(f'{col.title()} by Gender')
    plt.show()
    

    
```

4. Correlation Analysis

```python
# Correlation matrix
corr = df[['math score', 'reading score', 'writing score']].corr()
# Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Scores')
plt.show()
```

5. Insights & Recommendations

🔍 Key Insights
Test Preparation Improves Scores

Students who completed the test preparation course scored higher in math, reading, and writing on average.

Gender Differences in Performance

Females tend to score higher in reading and writing.

Males have a slight edge in math scores.

Lunch Type Affects Performance

Students with standard lunch performed better than those with free/reduced lunch — suggesting a possible socioeconomic factor.

Parental Education Level Correlates with Scores

Students whose parents have a higher level of education (Bachelor’s, Master’s) tend to score higher in all subjects.

Positive Correlation Between All Scores

Math, reading, and writing scores are strongly positively correlated — students who do well in one subject usually perform well in others.

💡 Recommendations
Encourage Test Preparation Programs

Since test preparation has a positive impact, schools should promote and provide access to such courses.

Support Students with Disadvantages

Extra academic support or mentoring should be considered for students on free/reduced lunch or from lower parental education backgrounds.

Focus on Gender-Inclusive Teaching

Address gender-based gaps by analyzing how teaching methods might be tailored to reduce those differences.

Integrated Learning Approaches

Because scores are correlated, cross-subject reinforcement (like combining reading and writing tasks) can improve overall performance.
